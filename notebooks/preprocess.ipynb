# Data Preprocessing


```python
import pandas as pd
import numpy as np
import datetime
import time
```

```python
original_dir = '../dataset/AStock_2021_2024.csv'
original_df = pd.read_csv(original_dir,encoding='utf-8')
original_df.columns
original_df.head(20)
```

```python
allstock_dir = '../dataset/CSI300_Stocks.xlsx'
AllStock = pd.read_excel(allstock_dir)
AllStock_code = list(AllStock['code'])
AllStock_name = list(AllStock['name'])
AllStock_code
```

```python
original_df['code'] = original_df['kdcode'].str.split('.').str[0]
AllStock_code_str = [f"{code:06d}" for code in AllStock_code]

set_original_code = set(original_df['code'])
set_allstock_code_str = set(AllStock_code_str)
intersection = set_original_code & set_allstock_code_str

indices = [i for i, code in enumerate(AllStock_code_str) if code in intersection]
set_AllStock_name = [AllStock['name'][i] for i in indices]
#print(len(set_AllStock_name))

filtered_original_df = original_df[original_df['code'].isin(intersection)]
filter_dir = '../dataset/filter_stock.csv'
filtered_original_df.to_csv(filter_dir,index=False,encoding='utf-8')
filtered_original_df
```

## Time Feature Processing


```python
filtered_original_df['time_pd'] = filtered_original_df['dt'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d'))
all_dates = set(filtered_original_df['time_pd'])
group_time = filtered_original_df.groupby('kdcode')

filtered_groups = []
for name, group in group_time:
    missing_count = len(all_dates) - group['time_pd'].nunique()
    if missing_count <= 100:
        filtered_groups.append(group)

filtered_original_df = pd.concat(filtered_groups)

filtered_original_df.shape
```

```python
def process_group(group):
    group_dates = set(group['time_pd'])
    missing_dates = all_dates - group_dates
    print(f"code: {group['kdcode'].iloc[0]}, missing dates: {len(missing_dates)}")
    missing_rows = pd.DataFrame(list(missing_dates), columns=['time_pd'])
    text_colums = ['close', 'open', 'high', 'low', 'prev_close','adjfactor', 'turnover', 'volume']
    for col in text_colums:
        if col != 'time_pd':
            missing_rows[col] = np.nan
    missing_rows['dt'] = missing_rows['time_pd'].apply(lambda x: x.strftime('%Y-%m-%d'))  
    missing_rows['kdcode'] = group['kdcode'].iloc[0]  
    missing_rows['code'] =  missing_rows['kdcode'].str.split('.').str[0]
    group = pd.concat([group, missing_rows], ignore_index=True)
    
    group = group.sort_values(by='time_pd')
    group = group.ffill().bfill()
    
    return group
```

```python
filtered_original_df = group_time.apply(process_group).reset_index(drop=True)
filtered_original_df['time_pd'] = filtered_original_df['dt'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d'))
```

```python
grouped_counts = filtered_original_df.groupby('kdcode').size() 
unique_counts = grouped_counts.unique()  

if len(unique_counts) == 1:  
    print(f"All rows are {unique_counts[0]}")  
else:  
    print(f"Each row is {unique_counts}") 
```

```python
noDuplicate = filtered_original_df.copy()
noDuplicate.shape
```

```python
if len(noDuplicate[['kdcode', 'time_pd']].drop_duplicates()) < len(noDuplicate):
    print("duplicate data exists!")
    noDuplicate= noDuplicate.drop_duplicates(subset=['code', 'time_pd'])
else:
    print("no duplicate data!")
noDuplicate.shape
```

```python
def process_group(group):
    missing_values = group.isnull().sum()
    print(f"Code: {group['code'].iloc[0]}   missing:{missing_values.sum()}")
    
    group = group.sort_values(by='time_pd')
    group = group.ffill().bfill()
    
    return group
```

```python
grouped = noDuplicate.groupby('kdcode')
noDuplicate_processed = grouped.apply(process_group).reset_index(drop=True)
noDuplicate_processed
```
